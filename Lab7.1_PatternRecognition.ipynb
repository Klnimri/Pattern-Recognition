```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate the data
num_hours_studied = np.array([1, 3, 3, 4, 5, 6, 7, 7, 8, 8, 10])
exam_score = np.array([18, 26, 31, 40, 55, 62, 71, 70, 75, 85, 97])

# Step 2: Visualize the data using scatter plots
plt.scatter(num_hours_studied, exam_score)
plt.title("Exam Score vs Number of Hours Studied")
plt.xlabel("Number of Hours Studied")
plt.ylabel("Exam Score")
plt.show()

# Step 3: Fit a linear regression using the normal equation and plot the results
X = np.c_[np.ones((num_hours_studied.shape[0], 1)), num_hours_studied]
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(exam_score)
y_predicted = theta[0] + theta[1] * num_hours_studied
plt.scatter(num_hours_studied, exam_score)
plt.plot(num_hours_studied, y_predicted, color="red")
plt.title("Exam Score vs Number of Hours Studied")
plt.xlabel("Number of Hours Studied")
plt.ylabel("Exam Score")
plt.show()

# Step 4: Predict the exam score for a student who studied for 9 hours
x_new = np.array([1, 9])
y_predicted_new = x_new.dot(theta)
print("The predicted exam score for a student who studied for 9 hours is:", y_predicted_new)
```
