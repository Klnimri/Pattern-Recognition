```python
#Task 2.1
import pandas as pd
import numpy as np

df = pd.read_csv('/kaggle/input/lab2-2/covidtotals.csv')
df.set_index('iso_code', inplace=True)

# Quantiles for cases column
print("Quantiles for Total Cases:")
print(np.arange(0, 1.1, 0.1))
print(df['total_cases'].quantile(np.arange(0, 1.1, 0.1)))

# Quantiles for deaths column
print("Quantiles for Total Deaths:")
print(np.arange(0, 1.1, 0.1))
print(df['total_deaths'].quantile(np.arange(0, 1.1, 0.1)))
```

```python
#Task 2.2
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/lab2-2/covidtotals.csv')
df.set_index('iso_code', inplace=True)

# Histogram of cases column
df['total_cases'].hist(bins=50)
plt.title('Histogram of Total Cases')
plt.xlabel('Number of Cases')
plt.ylabel('Frequency')
plt.show()
```

```python
#Task 3
#both distributions appear to be highly skewed to the right. 
#This means that the majority of the countries have relatively low numbers of cases or deaths, with a long tail of countries having much higher numbers.
```
